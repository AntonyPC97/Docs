```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('datosLimpios.csv',usecols=['Ciudad','Pais','PM25','PM10','O3','CO','NO2','Fecha'],parse_dates=['Fecha'],keep_date_col=True,dayfirst=True,date_parser=pd.to_datetime)
```

```python
df.head()
```

```python
df.sample(10)
```

```python
start = datetime.datetime(2018,1,1)
end = datetime.datetime(2018,1,31)
filtro = df.loc[(df['Fecha']>=start)&(df['Fecha']<=end)&(df['Ciudad']=='Connecticut')]
x = filtro['PM25']
plt.plot(x,label='PM 2.5')
x = filtro['PM10']
plt.plot(x,label='PM 10')
x = filtro['O3']
plt.plot(x,label='O3')
x = filtro['NO2']
plt.plot(x,label='NO2')
x = filtro['CO']
plt.plot(x,label='CO')

plt.legend(loc='best')
plt.title('Nivel AQI para los contaminanes en la ciudad Connecticut,United States\nEnero 2018')
plt.xlabel('Dia')
plt.ylabel('AQI Level')
plt.show()
```
